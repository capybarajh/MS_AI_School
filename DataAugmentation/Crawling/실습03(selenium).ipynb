```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By

import time
import os 
import urllib, requests
```

```python
### 1. query 선언 / chromedriver 실행
query = "딸기"
service = Service("./chromedriver.exe")
driver = webdriver.Chrome(service=service)

### 2. query 검색창에 추가 
driver.get("https://www.google.co.kr/imghp?h1=ko")
keyword = driver.find_element_by_xpath("/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/textarea")
keyword.send_keys(query)

# /html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/textarea -> 검색창
# /html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/button -> 검색창 버튼

### 3. 검색창에 입력어 들어오면 검색 실행
driver.find_element_by_class_name('Tg7LZd').send_keys(Keys.ENTER)

### 4. 스크롤 내려가기 -> 검색 더보기 클릭 후 -> 스크롤 내려가기
print(f"{query} 스크롤 내리는 중...")
elem = driver.find_element_by_tag_name('body')
for i in range(60) : 
    elem.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.1)
    
try : 
    driver.find_element_by_class_name('LZ4I').send_keys(Keys.ENTER)
    for i in range(60) : 
        elem.send_keys(Keys.PAGE_DOWN)
        time.sleep(0.1)
except :
    pass

### 5. 이미지 개수 파악하기
links = []
images = driver.find_elements_by_css_selector("img.rg_i.Q4LuWd")

for image in images:
    src = image.get_attribute('data-src') or image.get_attribute('data-iurl') or image.get_attribute('src')
    if src:
        links.append(src)

print("찾은 이미지 개수: ", len(links))

### 6. 이미지 다운로드
count = 0
for url in links:
    start = time.time()
    os.makedirs(f"./{query}_image_data/", exist_ok=True)
    while True:
        try:
            urllib.request.urlretrieve(url, f"./{query}_image_data/{str(count)}_{query}.png")
            print(f"{str(count + 1)} / {str(len(links))} / {query} / 다운로드 시간: {str(time.time() - start)[:5]} 초")
            break
        except urllib.error.HTTPError as e:
            print(f"HTTPError 발생 ({e}): 재시도 중...")
            time.sleep(5)
        except Exception as e:
            print(f"Error 발생 ({e}): 재시도 중...")
            time.sleep(5)
        if time.time() - start > 60:
            print(f"{query} 이미지 다운로드 실패")
            break
    count = count + 1

print(f"{query} 이미지 다운로드 완료")
driver.close()
```
