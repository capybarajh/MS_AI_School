# Numpy에서 많이 사용되는 함수

### 1. 정렬(기본 오름차순)

```python
import numpy as np
```

```python
array = np.array([5,20,30,2,4,8])
array.sort()
print(array)
```

### 2. 정렬(내림차순)

```python
print(array[::-1])
```

#### 원소의 정렬 - 각 열, 행을 기준으로 정렬

```python
array1 = np.array([[5,2,3,4,20], [7,9,10,4,45]])
print(array1)
org_array = array1.copy()
```

```python
# 각 열을 기준으로 정렬 전
print(array1)
```

```python
# 각 열을 기준으로 정렬 후
array1.sort(axis=0)
print(array1)
```

```python
# 각 행을 기준으로 정렬 후
org_array.sort(axis=1)
print(org_array)
```

## 균일한 간격으로 데이터 생성

# Numpy에서 많이 사용되는 함수

## 1. numpy 배열 객체 복사

```python
array1 = np.arange(0,10)
array2 = array1

array2[0] = 99

print(array2)
print(array1)
```

```python
# 복사
array3 = np.arange(0,10)
array4 = array3.copy()
array4[0] = 99

print(array3)
print(array4)
```

## 2. numpy 중복된 원소 제거

```python
array = np.array([1,1,2,2,3,3,4])
print("원소 값 :", array)
temp = np.unique(array)
print("중복 원소 처리 :", temp)
```

# 브로드캐스팅 개념 및 활용

```python
A = np.array([[1,2,3], [4,5,6]])
B = np.array([10, 20, 30])

# 브로드캐스팅을 통한 요소별 덧셈 연산
C = A + B
print(C)
```

## 이미지 저장

```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("./data/cat.png")

# 색상의 문제 발생 -> BGR -> RGB

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```

```python
import numpy as np
np.save("./data/image.npy", img)
```
