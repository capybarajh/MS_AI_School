# 기본적인 이미지 처리 기술을 이용한 이미지 선명도 처리

### 1. 기본 blur 처리

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 이미지 읽기 -> 그레이 스케일 변환
image = cv2.imread("./data/car.jpeg", 0)

plt.imshow(image, 'gray')
plt.show()
```

```python
blurred_1 = np.hstack([
    cv2.blur(image,(3,3)),
    cv2.blur(image,(5,5)),
    cv2.blur(image,(9,9)),
])

plt.imshow(blurred_1, 'gray')
plt.show()
```

### 2. Gaussian filter

```python
# 이미지 읽기 -> 그레이 스케일 변환
image2 = cv2.imread("./data/car.jpeg", 0)

gaussian_blurred_1 = np.hstack([
    cv2.GaussianBlur(image2, (3,3),0),
    cv2.GaussianBlur(image2, (5,5),0),
    cv2.GaussianBlur(image2, (9,9),0),
    
])
plt.imshow(gaussian_blurred_1, 'gray')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# 이미지 읽기 -> 그레이 스케일 변환
image = cv2.imread("./data/car.jpeg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

filter = np.array(([-1,-1,-1], [-1,9,-1], [-1,-1,-1]))

sharpne_img = cv2.filter2D(image, -1, filter)

plt.imshow(sharpne_img)
plt.show
```
