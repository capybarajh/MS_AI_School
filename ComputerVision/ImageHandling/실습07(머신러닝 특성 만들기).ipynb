# 머신러닝 특성 만들기

```python
import cv2
import matplotlib.pyplot as plt

image_path = "./data/box.png"
```

```python
image_gray = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

image_10x10 = cv2.resize(image_gray, (10,10)) # image 10x10 픽셀 크기로 변환
image_10x10.flatten()

plt.imshow(image_10x10, 'gray')
plt.show()
```

```python
image_color = cv2.imread("./data/apple.jpeg")
image_color = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)

image_color_10x10 = cv2.resize(image_color, (10,10)) # image 10x10 픽셀 크기로 변환
image_10x10.flatten().shape # 이미지 데이터를 1차원 벡터로 변환

# image size 255x255

image_color_255x255 = cv2.resize(image_color, (255,255))
image_color_255x255.flatten().shape # 이미지 데이터를 1차원 벡터로 변환

plt.imshow(image_color_10x10)
plt.show()

plt.imshow(image_color_255x255)
plt.show()
```
