```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(loc=0, scale=1, size=10000) # loc = 평균, scale = 표준편차, size = 크기
# print(data)

# 데이터 분포 시각화
sns.histplot(data)
plt.show()

# 데이터 통계 지표 계산
mean = np.mean(data)
std = np.std(data)
print("평균:", mean)
print("표준편차:", std)
```
