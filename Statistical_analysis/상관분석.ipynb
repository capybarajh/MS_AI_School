```python
import seaborn as sns
import pandas as pd 

tips_data = sns.load_dataset('tips')
print(tips_data)
```

```python
corr = tips_data[['total_bill', 'tip']].corr()
print(corr)
```

```python
sns.regplot(x='total_bill', y='tip', data=tips_data)
```

```python
# 피어슨 상관계수
corr = tips_data['total_bill'].corr(tips_data['tip'], method='pearson')
print(corr)
```

```python
# 공분산 실습 
covariance = tips_data['total_bill'].cov(tips_data['tip'])
print(covariance)

# 공분산이 클수록 두 변수가 상관관계 가짐, 0이면 두 변수는 독립관계
```

```python
# 산점도 실습 
sns.scatterplot(x='total_bill', y='tip', data=tips_data)
```
