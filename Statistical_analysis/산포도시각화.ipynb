```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 두 변수 간의 상관관계 생성
np.random.seed(0)
x = np.random.normal(loc=0, scale=1, size=100)
y = x + np.random.normal(loc=0, scale=0.5, size=100)
# print(y)

# 산포도 그리기
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Scatter Plot')
plt.show()
```
