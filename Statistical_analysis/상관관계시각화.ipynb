```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 데이터 생성 
np.random.seed(7777)
x = np.random.normal(loc=0, scale=1, size=100)
y = 2*x + np.random.normal(loc=0, scale=1, size=100)
print(y)
```

```python
# 산점도 시각화 
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Scatter Plot')
plt.show()
```

```python
# 상관관계 계산
corr = np.corrcoef(x,y)[0,1]
print("상관계수 :", corr)
```
