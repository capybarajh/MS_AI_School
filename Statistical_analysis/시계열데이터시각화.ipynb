# 시계열 데이터 탐색 및 시각화 

- 시계열 데이터 시각화 소개 (선 그래프, 박스 플롯, 자기상관 그래프)

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 시계열 데이터 생성 
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods=1000, freq='D')
values = np.random.randn(1000)

# 데이터 프레임 생성 
data = pd.DataFrame({
    'Date': dates, 
    'Value' : values
})

# 시계열 데이터 시각화 
plt.plot(data['Date'], data['Value'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Line Plot')
plt.xticks(rotation=-45)  
plt.show()
```

### 산점도 그래프 

```python
# 시계열 데이터 생성 
date_temp = pd.date_range(start='2023-01-01', periods=1000, freq='D')
val_temp = np.random.randn(1000)

# 프레임 생성
data_pd = pd.DataFrame({
    'Date' : date_temp, 
    'Vale' : val_temp
})

print(data_pd)
```

##### 산점도 그래프 시각화 

```python
plt.scatter(data_pd['Date'], data_pd['Vale'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Scatter Plot')
plt.xticks(rotation=-45)  
plt.show()
```

#### 막대 그래프 시각화 

```python
dates_bar_plot = pd.date_range(start='2023-01-01', periods=1000, freq='D')
val_bar_plot = np.random.randint(0,10,size=1000)

bar_plot_data = pd.DataFrame({
    'Date': dates_bar_plot, 
    'Value': val_bar_plot
})

print(bar_plot_data)
```

```python
plt.bar(bar_plot_data['Date'], bar_plot_data['Value'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Bar Plot')
plt.xticks(rotation=-45)  
plt.show()
```

#### 박스플롯 시각화

```python
box_date = pd.date_range(start='2023-01-01', periods=1000, freq='D')
box_val = np.random.randn(1000)

box_dp = pd.DataFrame({
    'Date': box_date,
    'Value' : box_val
})

print(box_dp)
```

```python
plt.boxplot(box_dp['Value'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Box Plot')  
plt.show()
```

#### 히스토그램 시각화

```python
histogram_date = pd.date_range(start='2023-01-01', periods=1000, freq='D')
histogram_val = np.random.randn(1000)

histogram_data = pd.DataFrame({
    'Date' : histogram_date,
    'Value' : histogram_val
})

print(histogram_data)

plt.hist(histogram_data['Value'], bins=20)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Time Series Histogram')  
plt.show()
```

#### 히트맵 시각화 

```python
import seaborn as sns

heat_date = pd.date_range(start='2023-01-01', periods=100, freq="D")
heat_val = np.random.randn(100)

heat_data = pd.DataFrame({
    'Date' : heat_date,
    'Value' : heat_val
})

print(heat_data)
```

```python
# 데이터를 피벗 테이블 변환 
pivot_table = heat_data.pivot(index='Date', columns='Value', values='Value')
# print(pivot_table)

# 히트맵 시각화 
plt.figure(figsize=(10,6))
sns.heatmap(pivot_table, cmap='YlOrRd', cbar=True)
plt.title('Time Series Heatmap')
plt.xlabel('Value')
plt.ylabel('Date')
plt.show()
```

피벗 테이블은 데이터를 특정 열을 기준으로 그룹화하여 요약하는 방법입니다.
