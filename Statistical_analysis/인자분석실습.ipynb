# 랜덤 데이터셋을 사용한 인자 분석 실습

```python
import numpy as np
from sklearn.decomposition import FactorAnalysis
```

```python
# 데이터셋 생성
n_samples = 100 # 샘플
n_feature = 5 # 변수 수

# 데이터셋을 랜덤하게 설정
x = np.random.randn(n_samples, n_feature)
print(x)
```

```python
# 인자 분석 수행
n_components = 2 # 추출할 인자 수
factorAnalysis = FactorAnalysis(n_components=n_components, random_state=0)
x_factor = factorAnalysis.fit_transform(x)

# 결과 출력
print("원본 데이터셋 크기", x.shape)
print(f"인자 분석 결과 인자 계수 : {n_components}")
print(x_factor)
```
