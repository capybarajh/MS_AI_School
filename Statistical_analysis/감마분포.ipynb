```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import gamma
```

```python
# 제품의 고장 까지 걸리는 시간 모델링 하는 감마분포 생성함수 
def generate_lifetime(alpha, beta, size) : 
    return np.random.gamma(shape=alpha, scale=beta, size=size) 

def measure_lifetiem(alpha, beta, num_samples) : 
    lifetime = generate_lifetime(alpha, beta, num_samples)
    
    
    plt.hist(lifetime, bins=30, density=True, alpha=0.8)
    plt.xlabel('Lifetime')
    plt.ylabel('Probability Density')
    plt.title('Product Lifetime Distribution')
    plt.show()
    
# 모수 설정
alpha = 2.5 # 감마분포 형상모수
beta = 1.2  # 감마분포 비율모수
num_samples = 1000
measure_lifetiem(alpha, beta, num_samples)
```
