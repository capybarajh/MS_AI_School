# 시계열 데이터 구성 요소 3가지(상향, 하향, 평평한 추세) 시각화

### 데이터 생성 (랜덤하게 시계열 데이터 생성)


```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
values = np.random.randn(100).cumsum()

# 데이터 프레임 생성
data = pd.DataFrame({
    'Date': dates, 
    'Value' : values
})

print(data)
```

## 상향 추세 시각화 

```python
data['Upward_Trend'] = np.linspace(0, 50, len(data))
print(data)
```

```python
plt.plot(data['Date'], data['Upward_Trend'], label="Upward_Trend")
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Upward Trend')
plt.legend()
plt.xticks(rotation=-45)  
plt.show()
```

```python
data['Downward_Trend'] = np.linspace(50,0, len(data))
print(data)
plt.plot(data['Date'], data['Downward_Trend'], label="Downward_Trend")
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Downward Trend')
plt.legend()
plt.xticks(rotation=-45)  
plt.show()
```

## 평평한 추세 시각화 

```python
data['Flat_Trend'] = np.full(len(data), 50)
print(data)
plt.plot(data['Date'], data['Flat_Trend'], label="Flat_Trend")
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Flat Trend')
plt.legend()
plt.xticks(rotation=-45)  
plt.show()
```
