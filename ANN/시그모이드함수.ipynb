# 인공지능 신경망 - 활성화 함수 시그모이드 실습 

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 시그모이드 함수

```python
def sigmaoid(x) : 
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 100) # -10 부터 10까지 100개 데이터 생성
print(x)
y = sigmaoid(x)

print(sigmaoid(100))

plt.plot(x, y)
plt.show()
```
