# 인공신경망 - 렐루 함수 실습 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def relu(x) : 
    return np.maximum(0, x)

x = np.linspace(-5, 5, 1000)
y = relu(x)

plt.plot(x, y)
plt.show()
```

# 인공신경망 -  Leaky ReLU 함수 실습

```python
def leaky_relu(x, alpha=0.01) : 
    return np.where(x >= 0, x, x * alpha)

x_temp = np.linspace(-5, 5, 100)
y_tmep = leaky_relu(x_temp)

plt.plot(x_temp,y_tmep)
plt.show()
```

# 인공신경망 - 소프트맥스 함수 

```python
def softmax(z) : 
    return np.exp(z) / np.sum(np.exp(z))

z = np.linspace(-10, 10, 100) # -10 부터 10까지 100개의 값을 생성하라 !!
y = softmax(z)

plt.plot(z, y)
plt.show()
```
