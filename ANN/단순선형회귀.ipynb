# 단순 선형 회귀의 개념 간단 실습

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,3,4,5,6])
print(x)
print(y)
```

```python
beta_1 = np.sum((x - np.mean(x)) * (y - np.mean(y))) / np.sum((x - np.mean(x))**2)
beat_0 = np.mean(y) - beta_1 * np.mean(x)
```

```python
print("기울기 : ", beta_1)
print("절편 : " , beat_0)
```

## 최소제곱법을 이용한 단순선형회귀 구현

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.show()
```

```python
# 회귀선 그리기
plt.scatter(x,y)
plt.plot(x, beat_0 + beta_1 * x, color='red')
plt.show()
```
