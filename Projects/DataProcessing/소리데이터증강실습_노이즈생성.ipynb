# Data Augmentation

- Adding noise: white noise 추가하여 원본 오디오에 잡음을 생성

```python
# !pip install soundfile
```

```python
import numpy as np
import soundfile as sf

# 오디오 듣기 위해서 사용할 라이브러리
import IPython.display as ipd
# ipd.Audio(y, rate=sr)
```

```python
# 샘플 Wav 파일 로드
file_path = "./data/blues.00001.wav"
# WAV : 일반적으로 Windows에서 사용되는 오디오 파일 형식 -> 고해상도 및 대용량 오디오 파일 저장하기 위해서 사용.
data, sr = sf.read(file_path)

# 원본 오디오 실행
ipd.Audio(data, rate=sr)
```

## 노이즈 생성

```python
# np.random.uniform() => 0과 1사이의 균등분포 값(랜덤한)
# np.amax() => 최댓값

noise_amp = 0.5*np.random.uniform()*np.amax(data) # 노이즈 크기를 조절하기 위한 스케일값
noise = noise_amp * np.random.normal(size=data.shape)

print("noise_amp: ", noise_amp)
print("noise: ", noise)
```

## 노이즈 추가

```python
data_noise = data.astype('float32') + noise.astype('float32')
print(data_noise)
```

## 노이즈가 추가된 음성 데이터 저장

```python
save_path = "./data/sample_noise_add.wav"
sf.write(save_path, data_noise, sr)
```

## 노이즈가 추가된 음성 읽기

```python
# 샘플 Wav 파일 로드
sample_wav = "./data/sample_noise_add.wav"

# WAV : 일반적으로 Windows에서 사용되는 오디오 파일 형식 -> 고해상도 및 대용량 오디오 파일 저장하기 위해서 사용.
sample_data, sample_sr = sf.read(sample_wav)

# 원본 오디오 실행
ipd.Audio(sample_data, rate = sample_sr)
```
