# MFCC - 간단한 실습

```python
import matplotlib.pyplot as plt

import librosa
import librosa.display
import numpy as np
```

## 1. 음성 파일 가져오기, SR 값 확인

```python
# 1. 음성 파일 가져오기
file_path = "./data/blues.00001.wav"
y, sr = librosa.load(file_path)
print("SR rate : ", sr)
```

## 2. 음성 파일에서 STFT 계산

```python
hop_length = 512 # 윈도우 이동크기 -> 스펙트로 그램을 생성할 때 얼마만큼 쪼개서 생성할거냐 ! 통상적으로 512 씀
n_fft = 2048 # FFT : 주어지느시간 대역에서 주파수 정보를 추출하는데 사용
stft = librosa.stft(y, n_fft=n_fft, hop_length=hop_length) # 시간-주파수 영역의 표현인 스펙트로그램 값

print(stft)
```

## 3. 멜 스케일 필터 방정식을 이용해서 Mel Spectrogram 계산

```python
n_mels = 40 # 20 ~ 40 사이 값으로 설정. 대부분의 음악분석은 20 ~ 40

# librosa.power_to_db() => 진폭을 데시벨 형태로 변환
# np.abs(stft)**2 => STFT 절댓값 제곱한 값
# STFT => 진폭을 나타내는 파워 스펙트로그램 계산 => mel_spec
mel_spec = librosa.feature.melspectrogram(S=librosa.power_to_db(np.abs(stft)**2), sr=sr, n_mels=n_mels)
print(mel_spec)
```

## 4. mel_spec ===> MFCC 계산하기

```python
n_mfcc = 13 # 음성 분야에서는 12 ~ 13 특정 벡터 추출하는 것이 일반적 ==> MFCC 원하는 개수의 특정 벡터를 추출하는 파라미터 값
mfccs = librosa.feature.mfcc(S=librosa.power_to_db(mel_spec), n_mfcc=n_mfcc)
print(mfccs)
```

## 5. 시각화 및 저장

```python
plt.figure(figsize=(10,4))
librosa.display.specshow(mfccs, x_axis='time')
plt.title('MFCC')
plt.tight_layout()

plt.savefig('./data/MFCC.png', dpi=300, bbox_inches='tight') # dpi = 해상도
# plt.close()

# dpi(dots per inch)는 인치 당 픽셀 수를 의미하며, 숫자가 클수록 더 고해상도의 이미지가 생성됩니다.
```
