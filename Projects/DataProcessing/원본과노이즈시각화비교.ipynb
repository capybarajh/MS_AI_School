# 원본 오디오와 잡음 생성된 오디오 비교

```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
```

## 원본 오디오 불러오기

```python
# 샘플 Wav 파일 로드
file_path = "./data/blues.00001.wav"
# WAV : 일반적으로 Windows에서 사용되는 오디오 파일 형식 -> 고해상도 및 대용량 오디오 파일 저장하기 위해서 사용.
data, sr = sf.read(file_path)
```

## 잡음 생성된 오디오 불러오기

```python
# 샘플 Wav 파일 로드
file_path = "./data/sample_noise_add.wav"

# WAV : 일반적으로 Windows에서 사용되는 오디오 파일 형식 -> 고해상도 및 대용량 오디오 파일 저장하기 위해서 사용.
noise_data, noise_sr = sf.read(file_path)
```

## 시각화

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.plot(data)
plt.title("Original")
plt.subplot(1,2,2)
plt.plot(noise_data)
plt.title("Noise Add")
plt.show()
```
