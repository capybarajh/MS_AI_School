```python
import matplotlib.pyplot as plt

import librosa
import librosa.display

import numpy as np

# 오디오 듣기 위한 라이브러리
import IPython.display as ipd

# 음성 파일 로드
filename = './data/blues.00001.wav'
y, sr = librosa.load(filename)
# y = 시계열 데이터, sr = 샘플레이트를 정수값으로 반환
#print(y, sr)

# 오디오
ipd.Audio(y, rate=sr) # rate=sr*2하면 속도 2배로 재생
```

```python
# 전처리
duration = 10 # 추출할 클립의 길이
samples = int(duration * sr) # 추출할 샘플 수
start = 0 # 추출 시작 시점(0초부터 시작)
y = y[start:start+samples]

# 스펙트로그램 생성
spec = librosa.stft(y)
spec_db = librosa.amplitude_to_db(np.abs(spec), ref=np.max)
```

```python
# 스펙트로그램 시각화
librosa.display.specshow(spec_db, sr=sr, y_axis='log', x_axis='time')
plt.title('Log-frequency power spectrogram')

# 이미지 파일로 저장
plt.savefig('./data/spectrogram.png')
#plt.close()
```
